## Analyse de forme

Le dataframe contient des informations sur le lieu de résidence des patients, notamment l'état et le code postal. 
L'Australie est une fédération constituée de 8 états, chacun étant défini par une plage de codes postaux (source: [Wikipedia](https://en.wikipedia.org/wiki/Postcodes_in_Australia)):
- **NSW (New South Wales):** 1000 -> 1999,  2000 -> 2599, 2619 -> 2899, 2921 -> 2999.
- **ACT (Australian Capital Territory) :** 	0200 -> 0299, 2600 -> 2618, 2900 -> 2920.
- **VIC (Victoria) :** 3000 -> 3999, 8000 -> 8999.
- **QLD (Queensland) :** 4000 -> 4999, 9000 -> 9999.
- **SA (South Australia) :** 5000 -> 5799, 5800 -> 5999.
- **WA (Western Australia) :** 	6000 -> 6797, 6800 -> 6999.
- **TAS (Tasmania) :** 7000 -> 7799, 7800 ->7999.
- **NT (Northern Territory):** 0800 -> 0899, 0900 -> 0999.

Le dataframe fournit également le nom, le prénom, l'âge et la date de naissance (au format AAAAMMDD) des patients. Ainsi qu'un numéro téléphone, qui est au format : **0X XXXXXXXX**. Il est constitué d'un code d'accès le '0' puis du code de la zone suivi du numéro local (source : [Wikipedia](https://en.m.wikipedia.org/wiki/Telephone_numbers_in_Australia)). 

- **00**: Accès international et d'urgence (voir ci-dessous les détails).
- **01**: Services téléphoniques alternatifs.
    - **014**: Téléphones satellites.
    - **0163**: Numéros de téléavertisseur.
    - **0198** Numéros de données.
- **02**: Région du Centre-Est ( NSW , ACT )
- **03**: Région du Sud-Est ( VIC , TAS )
- **04**: Services mobiles numériques ( 3G , 4G , 5G et GSM )
- **0550**: Services de communication indépendants de l'emplacement
- **07**: Géographie: région du Nord-Est ( QLD )
- **08**: Région du Centre et de l'Ouest ( SA , NT , WA )
- **1** Numéros non géographiques


Concernant la variable 'suburb', le gouvernement australien fournit des données permettant de faire le lien avec la ou les états correspondants, [GouvAus](https://www.abs.gov.au/websitedbs/D3310114.nsf/home/census) et le repository [git](https://github.com/michalsn/australian-suburbs).

```python
def scrapping_wikipedia(url):
    """
    This function find the poscode range by state from wikipedia.
    
    url = 'https://en.wikipedia.org/wiki/Postcodes_in_Australia'
    """    
    # open the url using urllib.request and put the HTML into the page variable
    page = urllib.request.urlopen(url)
    
    # parse the HTML from our URL into the BeautifulSoup parse tree format
    soup = BeautifulSoup(page, "lxml")
    
    # use the 'find_all' function to bring back all instances of the 'table' tag in the HTML and store in 'all_tables' variable
    all_tables=soup.find_all("table")
    
    # find the right table
    right_table=soup.find('table', class_='wikitable')
    
    # list of elements by column
    A=[]
    B=[]
    C=[]
    
    for row in right_table.findAll('tr'): # find rows
        cells=row.findAll('td') # find data
        if len(cells)==3: # 3 is the number of columns
            #append cell value
            A.append(cells[0].find(text=True)) 
            B.append(cells[1].find(text=True).lower()) 
            C.append(cells[2].get_text(separator=' ')) 
    
    # create the dataframe from table
    df=pd.DataFrame(A,columns=['State'])
    df['Abbreviation']=B
    df['Postcode range']=C
    
    # postcode range dictionnary
    poscode_range={}
    
    # for each state find the poscode range
    for abb in df.Abbreviation.values:
        poscode_range[abb] = re.findall('[0-9]+', 
                                             df['Postcode range'][df.index[df['Abbreviation'] == abb].tolist()[0]])
        
        
        poscode_range = {key:[int(i) for i in val] for key, val in poscode_range.items()}
        
    return poscode_range

dict_poscode_range= scrapping_wikipedia('https://en.wikipedia.org/wiki/Postcodes_in_Australia')
```

## Gestion des doublons

Le but de cet exercice étant de représenter et de discuter la prévalence de la maladie dans la population et non à des fins de prédiction, on peut alors ignorer les individus dont le patient_id n'est pas présent dans le dataframe _**df_pcr**_. Ce premier filtre permet de constater que le taux de données manquantes est identique entre le dataframe à 20000 lignes (df_patient initial) et le dataframe à 8800 lignes (df_patient après suppression des patient_id non présents dans le dataframe df_pcr). 

La méthode d'identification des doublons retenue ici passe par l'évaluation de la qualité des variables. Pour cela, on se concentre principalement sur les variables liées à la zone géographique, à savoir : state, phone number, suburb, postcode. Ainsi pour chaque patient, on va effectuer pour la variable:

- **State** : Un nettoyage pour se ramener à 1 des 8 états. Pour cela nous allons appliquer une recherche de correspondance pour chaque valeur de la colonne avec l'abréviation réelle des états.

- **Phone number**: On va récupérer le code de la zone et en déduire la/les état(s) correspondant(s).

- **Suburb**: On effectue la correspondance avec le dictionnaire _dict_suburb_.

- **Postcode**: On effectue la correspondance avec le dictionnaire _dict_poscode_range_ fournissant les plages en fonction des états.

Ces opérations permettent de résumer les informations portées par ces variables autour de la zone géographique (l'état). Ainsi, pour chaque individu on évalue l'occurence de chaque état à laquelle on applique une pondération. La pondération provient de l'évaluation du taux de patient par état en partant de la variable présentant le moins de données manquantes, à savoir _post_code_. On observe que les pondérations obtenues sont fortement corrélées avec les densités réelles (source : [Worldatlas](https://www.worldatlas.com/articles/the-largest-states-and-territories-of-australia.html)).

Cette méthode va non seulement permettre d'identifier l'état le plus probable. Mais aussi, pour des  patients présentant le même patient_id récupérer celui présentant la plus grande probabilité d'appartenir à un état parmi ces candidats.

```python
df_patient_cln = df_patient.copy()

# subset only patient_id in df_pcr
df_patient_cln = df_patient_cln[df_patient_cln.patient_id.isin(df_pcr.patient_id.unique())]
```

```python
# Missing values and gap between df_patient init
total = df_patient_cln.isnull().sum().sort_values(ascending = False)
percent = round(total/df_patient_cln.shape[0]*100 ,2)
dif = percent - missing_values['%']
pd.concat([total, percent, dif], axis = 1, keys=["Total missing values","%", "Gap"])
```

```python
# extract area zone from phone number 
df_patient_cln['area'] = df_patient_cln.phone_number.str.extract(r'([^\s]+)')
```

```python
# switch wrong values in columns suburb and postcode

# index of wrong column values  
idx_swap = df_patient_cln[df_patient_cln.postcode.str.isdigit() == False].index.tolist()
    
# swap wrong columns
df_patient_cln.loc[idx_swap, ['suburb', 
                              'postcode']] = df_patient_cln.loc[idx_swap, ['postcode', 
                                                                 'suburb']].values
```

```python
def state_data_cleaning(df, state_dict):
    """ 
    This function return the right state's name.
    
    p:
        df : the dataframe
        state_dict : the corresponding state based on sequences comparison
    """
    # edit dataframe columns state
    if df['state'] is None:
        return None
    else:
        return state_dict[df['state']]
```

```python
# add cleaned states column
df_patient_cln['state_cln'] = df_patient_cln.apply(lambda x: state_data_cleaning(x, 
                                                                                 dict_all_states),
                                                   axis= 1)

df_patient_cln.drop(['state'], axis= 1, inplace= True)
```

```python
def match_postcode(df, post_dict):
    """
    This function return the state for given postcode.
    
    p: 
        df : the dataframe
        post_dict : the corresponding state based on postcode 
    """
    
    if df['postcode'] is None:
        return None
    else:
        for ps_state in post_dict.keys():
            groups= list(zip(*[iter(post_dict[ps_state])]*2))
            if any(start <= int(df['postcode']) <= end for start,end in groups):
                return ps_state
```

```python
# add state from postcode
df_patient_cln['postcode_cln'] = df_patient_cln.apply(lambda x: match_postcode(x, dict_poscode_range),
                                  axis= 1)

df_patient_cln.drop(['postcode'], axis= 1, inplace= True)
```

```python
# weight by state
percent_wgt = df_patient_cln.postcode_cln.value_counts()/df_patient_cln.postcode_cln.value_counts().sum()

wgt = percent_wgt.to_dict()
```

```python
def zone_name(df, phone_dict):
    """
    The function return the states fo each area zone
    
    p: 
        - df : the dataframe
        - phone_dict : the corresponding states based on area code
    """

    if df['area'] in phone_dict.keys():
        return phone_dict[df['area']]
    else:
        return None
```

```python
df_patient_cln['area_cln'] = df_patient_cln.apply(lambda x: zone_name(x, dict_phone_number),
                                  axis= 1)
df_patient_cln.drop(['area'], axis= 1, inplace= True)
```

```python
def match_suburb(df, suburb_dict):
    """
    This function return the states fo each suburb
    
    p: 
        - df : the dataframe
        - suburb_dict : the corresponding states based on suburb name
    """
    if df['suburb'] is None:
        return None
    elif df['suburb'] in suburb_dict.keys():
        return ' '.join(suburb_dict[df['suburb']])
    else:
        return None
```

```python
df_patient_cln['suburb_cln'] = df_patient_cln.apply(lambda x: match_suburb(x, dict_suburb),
                                  axis= 1)

df_patient_cln.drop(['suburb'], axis= 1, inplace= True)
```

```python
def detect_duplicates(df):
    """
    This function detect duplicated patient
    
    p: 
        - wgt: density by state
    """
    
    # list of duplacate index
    idx_drop = []
    
    # list of unique patient_id
    unique = df.patient_id.unique()
    
    # list of patient_ids having more than 1 occurence
    dpt_patient_id= []
    for id_pat in unique:
        if len(df[df.patient_id == id_pat]) > 1:
            dpt_patient_id.append(id_pat)
    
    # for each patient id duplicated
    for row in dpt_patient_id:
        check_right = df[df.patient_id == row]
        
        # dictionnary for idx et state max.
        dict_max_idx = {}
        dict_max_state = {}
        
        for row in range(len(check_right)):
            # list of states for each row 
            list_state = ' '.join([i for i 
                                   in check_right.iloc[row]['state_cln': 
                                                            'suburb_cln'].values.tolist() if i]).split(' ')
            
            # occurance for each state in the list 
            count = dict(Counter(list_state))
            
            # add weight to count.
            for cnt in count.keys():
                count[cnt] = count[cnt]*wgt[cnt]
            
            # find index of max ans the state.
            dict_max_idx[check_right.index[row]] = max(count, key=count.get)
            dict_max_state[check_right.index[row]] = max(count.values())
            
            
        # add the index of row to drop    
        idx_drop.append(min(dict_max_state, key=dict_max_state.get))
    
    # drop duplicated rows
    df.drop(idx_drop, inplace= True)
    
    return df
```

## Jointure avec df_pcr

La jointure est effectuée en utilisant le dataframe après suppression des doublons. Le dataframe df_pcr contenant des patient_id en double on va se retrouver face à des incohérences, notamment si pour un même patient_id on a deux résultats différents. 



On vérifie alors pour chaque patient_id en double, si le résultat est identique. Si oui on ne garde qu'un seul index, si non on supprime les 2 index et donc le patient_id.



De plus, on ajoute une colonne **Finale_State** qui permet de résumer l'information portée par les colonnes : states, postcode, suburb et phone number. Pour cela on applique les même fonctions qui y'ont permis de  supprimer les doublons mais cette fois-ci on cherche l'état le plus probable. 

```python
dataset = df_patient_cln.merge(df_pcr)
```

```python
df_patient_pcr = dataset.copy()
```

```python
def encoding_pcr(df):
    """
    This function encode the column pcr
    """
    if df['pcr'] == 'N':
        return 'Negative'
    elif df['pcr'] == 'P':
        return 'Positive'
    else:
        return df['pcr']
```

```python
df_patient_pcr['Test'] = df_patient_pcr.apply(encoding_pcr, axis= 1)
df_patient_pcr.drop(['pcr'], axis= 1, inplace= True)
```

```python
# for each patient_id give the most probable state
max_state_wgt = []

for row in range(len(df_patient_pcr)):
    list_state = ' '.join([i for i 
                           in df_patient_pcr.iloc[row]['state_cln': 
                                                    'suburb_cln'].values.tolist() if i]).split(' ')
    count = dict(Counter(list_state))
    
    for cnt in count.keys():
        count[cnt] = count[cnt]*wgt[cnt]
        
    max_state_wgt.append(max(count, key=count.get))

df_patient_pcr['Final_State'] = max_state_wgt
```

```python
df_patient_pcr.head()
```

```python
def drop_dpt_pcr(df):
    """
    This function return a unique row for patient_id having more than one index
    """
    drop_ixd = []
    
    # get duplicated patient_id
    dpt_patient_id= []
    for id_patient in df.patient_id.unique():
        if len(df[df.patient_id == id_patient]) > 1:
            dpt_patient_id.append(id_patient)
    
    for row in dpt_patient_id:
        check_df = df[df.patient_id == row]
        if check_df.iloc[0, df.columns.get_loc('Test')] != check_df.iloc[0, df.columns.get_loc('Test')]:
            drop_ixd.extend(check_df.index.tolist())
        else:
            drop_ixd.extend([check_df.index.tolist()[0]])
    
    return drop_ixd
```

```python
drop_idx_pcr = drop_dpt_pcr(df_patient_pcr)
df_patient_pcr.drop(drop_idx_pcr, inplace=True)
```
